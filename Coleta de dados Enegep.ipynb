```python
# paarte 2 - definindo a url
pg1 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2013&area=177&pchave=&autor=&categ='
pg2 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2013&area=178&pchave=&autor=&categ='
pg3 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2017&area=240&pchave=&autor=&categ='
pg4 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2017&area=241&pchave=&autor=&categ='
pg5 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2017&area=242&pchave=&autor=&categ='
pg6 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2017&area=243&pchave=&autor=&categ='
pg7 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2017&area=244&pchave=&autor=&categ='
pg8 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2017&area=245&pchave=&autor=&categ='
pg9 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2017&area=246&pchave=&autor=&categ='
pg10 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2017&area=247&pchave=&autor=&categ='
pg11 = 'https://www.abepro.org.br/publicacoes/index.asp?pesq=ok&ano=2017&area=248&pchave=&autor=&categ='


site = requests.get(pg2)

soup = BeautifulSoup(site.content, 'html.parser')
artigos = soup.find_all('div', class_='div_trabalho')
```

```python
# parte 3 - criando a função para extrair os dados e rodando o programa
def artigo_dic(tema_artigo, subtema, numero_artigo, titulo, autores, qtd_autores, palavras_chaves, qtd_palavras_chaves, link, resumo):
    dic_art = {
        'TEMA ARTIGO':tema_artigo,
        'SUBTEMA': subtema,
        'NUMERO ARTIGO':numero_artigo,
        'TITULO': titulo, 
        'AUTORES': autores,
        'QTD AUTORES': qtd_autores,
        'PALAVRAS CHAVES': palavras_chaves,
        'QTD PALAVRAS CHAVES': qtd_palavras_chaves,
        'LINK': link,
        'RESUMO': resumo        
    }
    return dic_art

dados_artigos = []

for num in range(0, len(artigos)):
    
    artigo = artigos[num].text.strip().split('\n')
    
    print(100*'-')
    print(f'Artigo: {num+1} \n') # Quantidade de Artigos
    tema_artigo = artigos[num].find('font').text.strip()[:] # Tema do Artigo
    print(f'Tema do Artigo: ', tema_artigo, '\n')
    
    subtemas = artigos[num].text.split('\n')
    subtema = subtemas[6].split('/')[1][:].strip() # Subtema do Artigo
    print(f'Subtema do Artigo: ', subtema, '\n')
    
    numero_artigo = artigo[0].strip() # Número do Artigo
    print(f'Número do Arigo: ', numero_artigo, '\n')
    
    titulo = artigo[2].strip() # Título do Artigo
    print(f'Título o Artigo: ', titulo, '\n')
    
    autores = artigos[num].find('b', string='AUTORES:').next_sibling.strip() # Autores do Artigo
    print(f'Autores do Artigo: ', autores, '\n')
    print(f'Quantidade de Autores : ', len(autores.split(';')), '\n')
    
    palavras_chaves = artigos[num].find_all('small') # procurando as palavras-chaves do artigo, tem 0 e 1
    palavras = palavras_chaves[1].get_text().strip().upper() # Palavras-chaves do Artigo
    
    if len(palavras.split(',')) > len(palavras.split(';')):
        
        quantidade_palavras = len(palavras.split(','))
        
    else:
        quantidade_palavras = len(palavras.split(';'))

    print(f'Palavras-chave do artigo: ', palavras, '\n') # Palavras-chaves do Artigo
    print(f'Quantidade de Palavras-chave: ', quantidade_palavras, '\n') # Quantidade de Palavras-chaves do Artigo
    
    link_artigo  = artigos[0].find_all('meta') # Link do Artigo
    link = link_artigo[1].get('content').strip()
    print(f'Link para download: ', link, '\n')
    
    resumo = artigo[14].strip().split('  ').pop(0) # Resumo do Artigo
    print(f'Resumo do Artigo: ', resumo)
    print(100*'-')
    dados_artigo = artigo_dic(tema_artigo, subtema, numero_artigo, titulo, autores, len(autores.split(';')), palavras, quantidade_palavras, link, resumo)
    
    dados_artigos.append(dados_artigo)
    #if num == 2:
        #break
```

```python
# parte 4 - criando o dataframe
df = pd.DataFrame(dados_artigos)
df.head()
```

```python
# parte 5 - exportando para excel, mude o nome do arquivo para não sobrescrever o anterior exemplo: Artigos1.xlsx, Artigos2.xlsx, etc.
# Basta mudar a numeração
df.to_excel('./Arquivo artigos/Arigos 2017/Artigos11.xlsx', index=False)
```

```python
# parte 7 - juntando todos os arquivos em um só
# detalhe importante, todos os arquivos devem estar na mesma pasta e mude o nome que está aqui no programa
# exemplo: #caminho = os.path.abspath('./nome_do_diretorio')
import os
# Para juntar todos os arquivos diários baixados no site da magalu
# caminho dos arquivos
caminho = os.path.abspath('./Arquivo artigos/Arigos 2017')
# defiição do dataframe vazio
dfvazio = pd.DataFrame()

print(f'Mostrando o tamaho do corpo do dataframe vazio: {dfvazio.shape}')

for num, file in enumerate(os.listdir(caminho)):
    # o for para iterar sobre os caminhos dos arquivos
    # juntando o caminho com o nome do arquivo
    path = os.path.join(caminho, file)
    # leitura do arquivo
    dfs = pd.read_excel(path)
    # concatenação dos arquivos
    dfvazio = pd.concat([dfvazio, dfs], ignore_index=True)
    # mostra o tamanho do arquivo atual e o acumulado
    print(f'Arquivo {num+1} - {file} - Tamanho do arquivo atualmete: {dfs.shape} - arquivo acumulado: {dfvazio.shape}')
    print(100*'-')
# resultado final    
print(f'Tamanho do aquivo final: {dfvazio.shape}')

dfvazio.to_excel('./Arquivo artigos/Arigos 2017/Artigos_final_2017.xlsx', index=False)

print(type(dfvazio))
df = pd.read_excel('./Arquivo artigos/Arigos 2017/Artigos_final_2017.xlsx')
```
